Build a prediction model for Salary_hike

```python
import pandas as pd

df = pd.read_csv("D:/Desktop/Assignments/Assignment4/Salary_Data.csv")
df.head()
```

```python
import seaborn as sns

sns.distplot(df['Salary'],color='orange')
```

```python
sns.distplot(df['YearsExperience'],color='black')
```

```python
x = df.iloc[:,:-1]
x
y = df.iloc[:,-1]
y
```

From the Graph we can see as the yearExperience increase the salary also get increase as we can say both are directly proportional

```python
sns.scatterplot(x = df['Salary'],y = df['YearsExperience'],data=df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train,x_test,y_train,y_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#training Accuracy

y_pred_train = model.predict(x_train)
y_pred_train
print("R2_Score",r2_score(y_train,y_pred_train))
print("mean_absolute_error",mean_absolute_error(y_train,y_pred_train))
```

```python
# Testing Accuracy

y_pred_test = model.predict(x_test)
y_pred_test
print("r2_score",r2_score(y_test,y_pred_test))
print("mean_absolute_error",mean_absolute_error(y_test,y_pred_test))
```

```python
sns.regplot(x = y_test,y = y_pred_test)
```

```python
sns.regplot(x = y_train,y = y_pred_train)
```
