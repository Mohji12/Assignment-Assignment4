```python
# Predicting Delivery Time on the Bases of Soritng Time

import pandas as pd
```

```python
df = pd.read_csv("D:/Desktop/Assignments/Assignment4/delivery_time.csv")
df.head()
```

# In above dataset we have two features Delivery time and Sorting and total number data points is 21
# Sorting Time is the Independent variable and Delivery Time is Dependent Variable

```python
# Distribution of Each Columns

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(df['Delivery Time'],bins=5,color='green')
plt.show()
```

```python
sns.distplot(df['Sorting Time'],bins=5,color='purple')
plt.show()
```

# from Above Graph we can say that data is in normal form

```python
x = df.iloc[:,1:2]
x
y = df.iloc[:,:-1]
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.2)
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Testing Accuracy
y_pred = model.predict(x_test)
y_pred

r2_score(y_test,y_pred)
```
